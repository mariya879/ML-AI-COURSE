PLOTTING EACH COLUMN OF THE DATA FOR IDENTIFYING THE DISTRIBUTION OF THE DATA

```python
# Plot histograms for numerical columns and count plots for categorical columns
for col in df.columns:
    if df[col].dtype == 'float64' or df[col].dtype == 'int64':
        plt.figure(figsize=(8, 6))
        sns.histplot(df[col], bins=30, kde=True)
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()
    else:
        plt.figure(figsize=(8, 6))
        sns.countplot(x=col, data=df)
        plt.title(f'Count Plot of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.show()
```

FOR CATEGORICAL VALUE MODE IS SUITABLE METHOD FOR FILLING THE NAN VALUES

```python
df['Outcome'].fillna(df['Outcome'].mode().iloc[0], inplace=True)  # Fill NaN with mode of 'Outcome' column
```

FOR NUMERICAL VALUES

IF THE DISTRIBUTION IS LEFTLY SKEWED REPLCE WITH MEAN AND IF RIGHT REPLCE WITH MEDIAN 

```python

numerical_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

```

```python
for col in numerical_cols:
    df[col].fillna(df[col].mean(), inplace=True)  # Fill NaN with median
   
print(df.isnull().sum())
```
